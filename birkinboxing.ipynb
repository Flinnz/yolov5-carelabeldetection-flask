```python
def count_files(dir_name):
    count = 0
    for path in os.listdir(dir_name):
        if os.path.isfile(os.path.join(dir_name, path)):
            count += 1
    return count


pic_dir = 'pics'
for path in os.listdir(dir_name):
    if os.path.isdir(os.path.join(pic_dir, path)):
        count = count_files(os.path.join(pic_dir, path))
        if count < 6:
            print(path, count)
```

```python
import pandas as pd
```

```python
annotations = ['carelabel/Carelabel128/annotation', 'carelabel/CL2018/CL2018_eval/annotation', 'carelabel/CL2018/CL2018_train/annotation']
df = pd.DataFrame(
    {
        "x1": [],
        "y1": [],
        "x2": [],
        "y2": [],
        "x3": [],
        "y3": [],
        "x4": [],
        "y4": [],
        "type": [],
        "value": [],
        "file_name": [],
        "img_file": [],
    }
    
)
df.columns
```

```python
for annotation_folder in annotations:
    img_folder = annotation_folder[:-len('annotation')]
    for path in os.listdir(annotation_folder):
        file_path = os.path.join(annotation_folder, path)
        if os.path.isfile(file_path):
            n_df = pd.read_csv(file_path, sep=',', names=df.columns)
            n_df['file_name'] = os.path.join(annotation_folder, path[3:])
            n_df['img_file'] = os.path.join(img_folder, path[3:][:-4] + '.jpg')
            df = pd.concat([df, n_df])
df = df[df.type == 'Carelabel']
df
```

```python
classes = {
    0: ['washing_normal_40', 'washing_not', 'washing_mild_60',
                'washing_normal_30', 'washing_veryMild_30', 'washing_mild_30',
                'washing_normal_60', 'washing_mild_40', 'washing_hand_40', 'washing_veryMild_40'], # washing
    1: ['bleaching_not', 'bleaching_onlySome', 'bleaching_any'], #bleaching
    2: ['professional_dry_not', 'professional_dry_normal_P',
                         'professional_dry_mild_P', ], # cleaning
    3: ['drying_tumble_not', 'drying_tumble_lower_60',
               'drying_tumble_normal_80', 'drying_line', 'drying_flat' ], # dry
    4: ['ironing_plate_110', 'ironing_plate_150',
                'ironing_not', 'ironing_plate_200'], # ironing
}
classes
```

```python
classes_inverse = {}
for k in classes.keys():
    for cls in classes[k]:
        classes_inverse[cls] = k
classes_inverse
```

```python
df.value = df.value.map(classes_inverse)
```

```python
from PIL import Image
```

```python
def get_image_size(img_path):
    img = Image.open(img_path)
    return int(img.size[0]), int(img.size[1])
```

```python
df['image_width'], df['image_height'] = zip(*df.img_file.map(lambda img: get_image_size(img)))
df
```

```python
df['xmax'] = df[['x1', 'x2', 'x3', 'x4']].max(axis=1)
df['xmin'] = df[['x1', 'x2', 'x3', 'x4']].min(axis=1)
df['ymax'] = df[['y1', 'y2', 'y3', 'y4']].max(axis=1)
df['ymin'] = df[['y1', 'y2', 'y3', 'y4']].min(axis=1)
```

```python
df
```

```python
def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return [x,y,w,h]
```

```python
df['x'], df['y'], df['w'], df['h'] = zip(*df.apply(lambda x: convert((x['image_width'], x['image_height']), [x['xmin'], x['xmax'], x['ymin'], x['ymax']]), axis=1))
```
